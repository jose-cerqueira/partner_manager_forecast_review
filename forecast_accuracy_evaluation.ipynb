Hypothesis Formulation and Testing
1. The level of significance (Alpha) = 0.05
2. The sample size N = 48 which is sufficiently large for a Z stat Test.
3. But since the population standard deviation (Sigma) is unknown, we have to use a T stat Test.
4. Since the sample is different for both Sampling tests, we have N+N-2 degrees of freedom = 94
5. Since the sole purpose of the test is to check whether there is any effect on driving distances due to the new coating, we could prefer a Two Tailed T Test.

Hypothesis Formulation:

Use two tailed independent sample T test for means

Null Hypothesis:

H0:fornfor_rev=0 (Reviewing the forecast does not have effect on error reduction)

Alternate Hypothesis:

H1:forfor_rev0 (Reviewing the forecast does have significant effect on error reduction)

## Welch's T-test

```python
def welch_dof(x,y):
        dof = (x.var()/x.size + y.var()/y.size)**2 / ((x.var()/x.size)**2 / (x.size-1) + (y.var()/y.size)**2 / (y.size-1))
        print(f"Welch-Satterthwaite Degrees of Freedom= {dof:.4f}")
        return np.around(dof,decimals=4)
        
def welch_ttest(x, y): 
    ## Welch-Satterthwaite Degrees of Freedom ##
    dof = (x.var()/x.size + y.var()/y.size)**2 / ((x.var()/x.size)**2 / (x.size-1) + (y.var()/y.size)**2 / (y.size-1))
   
    t, p = stats.ttest_ind(x, y, equal_var = False)
    
    print("\n",
          f"Welch's t-test= {t:.4f}", "\n",
          f"p-value = {p:.4f}", "\n",
          f"Welch-Satterthwaite Degrees of Freedom= {dof:.4f}")
```

```python
print("NS :")
ns_dof=welch_dof(data_abs['ns_error'],data_abs['ns_rev_error'])
print("WI :")
wi_dof=welch_dof(data_abs['wi_error'],data_abs['wi_rev_error'])
print("SOS1D :")
sos1d_dof=welch_dof(data_abs['sos1d_error'],data_abs['sos1d_rev_error'])
print("SOS2D :")
sos2d_dof=welch_dof(data_abs['sos2d_error'],data_abs['sos2d_rev_error'])
```

```python
welch_ttest(data_abs['ns_error'],data_abs['ns_rev_error'])
```

```python
welch_ttest(data_abs['wi_error'],data_abs['wi_rev_error'])
```

```python
welch_ttest(data_abs['sos1d_error'],data_abs['sos1d_rev_error'])
```

```python
welch_ttest(data_abs['sos2d_error'],data_abs['sos2d_rev_error'])
```

## T test Conclusion
* In all metrics, the p value is greater than the 0.05.
* Hence, we failed to reject the Null Hypothesis, and there is no significant difference between the means of Forecast Error and Forecast Review Error
